```python
import pandas as pd

# Loading dataset
df = pd.read_csv("https://catalog.ourworldindata.org/garden/covid/latest/compact/compact.csv")  # replace with the downloaded CSV file name

# Data preview 
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Converting date to datetime
df['date'] = pd.to_datetime(df['date'])

# Droping rows with missing critical data
df = df.dropna(subset=['total_cases', 'total_deaths'])

# Filtering for a few countries (example: top 5 by total cases)
top_countries = df.groupby('country')['total_cases'].max().sort_values(ascending=False).head(5).index
df_top = df[df['country'].isin(top_countries)]

df_top.head()
```

```python
import matplotlib.pyplot as plt

# Plotting Daily new cases trend for top 5 countries
plt.figure(figsize=(12,6))
for country in top_countries:
    country_data = df_top[df_top['country'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.title('Daily New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('Cases')
plt.legend()
plt.show()
```

```python
# Saving cleaned dataset
df_top.to_csv('covid_top_countries_cleaned.csv', index=False)
```

```python
# Vaccination progress analysis
plt.figure(figsize=(12,6))
for country in top_countries:
    country_data = df_top[df_top['country'] == country]
    plt.plot(country_data['date'], 
             country_data['people_fully_vaccinated_per_hundred'], 
             label=country)

plt.title('COVID-19 Vaccination Progress (Fully Vaccinated %)')
plt.xlabel('Date')
plt.ylabel('% of Population Fully Vaccinated')
plt.legend()
plt.show()
```

```python
# Assuming you already have df loaded from the CSV
df_countries = df[~df['country'].isin([
    'World', 'Africa', 'Asia', 'Europe', 'European Union',
    'High income', 'Low income', 'Lower middle income', 'Upper middle income', 'Oceania'
])]
```

```python
top_countries = df_countries.groupby('country')['total_cases'].max().nlargest(5).index
```

```python
df_top = df_countries[df_countries['country'].isin(top_countries)].copy()
```

```python
# calculating mortality rate
df_top['mortality_rate'] = (df_top['total_deaths'] / df_top['total_cases']) * 100
```

```python
plt.figure(figsize=(12,6))
for country in top_countries:
    country_data = df_top[df_top['country'] == country]
    plt.plot(country_data['date'], 
             country_data['mortality_rate'], 
             label=country)

plt.title('COVID-19 Mortality Rate Over Time (Top 5 Countries)')
plt.xlabel('Date')
plt.ylabel('Mortality Rate (%)')
plt.legend()
plt.show()
```

```python
summary_table = df_top.groupby('country').agg(
    Max_Cases=('total_cases', 'max'),
    Max_Deaths=('total_deaths', 'max'),
    Max_Vaccinations=('total_vaccinations', 'max'),
    Mortality_Rate_Last=('mortality_rate', 'last')
).reset_index()

# Sort by Max Cases
summary_table = summary_table.sort_values(by='Max_Cases', ascending=False)
```

```python
import pandas as pd

# Ensure columns are numeric by removing commas and converting
for col in ['Max_Cases', 'Max_Deaths', 'Max_Vaccinations']:
    summary_table[col] = (
        pd.to_numeric(
            summary_table[col]
            .astype(str)       # make sure everything is a string
            .str.replace(',', ''),  # remove commas
            errors='coerce'    # turn invalid stuff into NaN
        )
    )

# Fill NaN with 0
summary_table = summary_table.fillna(0)

# Convert to integers where applicable
summary_table[['Max_Cases', 'Max_Deaths', 'Max_Vaccinations']] = summary_table[['Max_Cases', 'Max_Deaths', 'Max_Vaccinations']].astype(int)

# Keep a numeric table for calculations
summary_numeric = summary_table.copy()

# Create display table with formatting
summary_display = summary_table.copy()
summary_display['Max_Cases'] = summary_display['Max_Cases'].apply(lambda x: f"{x:,}")
summary_display['Max_Deaths'] = summary_display['Max_Deaths'].apply(lambda x: f"{x:,}")
summary_display['Max_Vaccinations'] = summary_display['Max_Vaccinations'].apply(lambda x: f"{x:,}" if x > 0 else "N/A")
summary_display['Mortality_Rate_Last'] = summary_display['Mortality_Rate_Last'].apply(lambda x: f"{x:.2f}%" if x > 0 else "N/A")

summary_display
```

```python
# Converting DataFrame to Markdown table
markdown_table = summary_display.to_markdown(index=False)

# Save it to a file (optional)
with open("covid_summary_table.md", "w") as f:
    f.write(markdown_table)

print(markdown_table)
```
